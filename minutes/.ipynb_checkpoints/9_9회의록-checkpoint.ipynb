### 9.9(수) 화상회의
#### 1. 첫번째 회의(11:00~12:30)

#### 내용

- 한계점
 1. 데이터 : 여러가지 특성을 가진 데이터를 구하기가 힘들다.
 2. 데이터의 특성들이 몇개 없다보니 있는 특성으로는 매운것, 단것 등등을 구별하기엔 어려움이 있다.

- 고민점
 1. 음식을 구별할 특성을 찾기.
 2. 음식 추천의 새로운 방법 모색
 3. 폐쇄형 설문조사로 음식후보군을 지정해서 설문받기 등등 .....

#### 대안
1. 이희진 - 설문조사를 성별, 연령, 기분(1~10)에 따라 먹고싶은 음식을 먹는다는 것을 수치로 설문 받아 협업필터링을 이용하여 음식 추천
2. 유세빈 
 - 역으로 스트레스가 풀리는 재료를 이용하여 음식 추천하기. ex) 화가날 때에는 양파. 
 - 분류 : 스트레스 / 짜증 / 화 / 우울 / 허전함 / 기쁠때 / 피로할 때 / 무기력할 때 등
 
#### 결과
1. 크롤링한 데이터를 이용하여 기분별 재료를 기반으로 추천해보기. 
2. 배운 추천알고리즘을 이용할 수 있는 방향으로 구현할 수 있는 기능 생각해오기.

### 2. 두번째 회의(16:00~18:00)

이희진  
- 클러스터링을 이용한 협업필터링을 구현해보자.  

최재현  
- 기분전환을 위한 음식 추천.  
- ex) 먹으면 기분이 좋아지는 음식을 표본조사를 기반으로 추천.  

서예지  
- 연령별, 성별, 각 카테고리별 얼마나 어울리는 음식을 받는것.  

유세빈
- 배운 협업필터링을 기반으로 생각을 해봤는데 한계가 있었다.

#### 결과
- 사용자 기반 유사도 각 카테고리별(나쁠 때 / 좋을 때) 요리 점수를 이용하여
- 설문조사시 요리점수를 받을 방법을 고안해보자 다양한 카테고리(찌개류,튀김류,면류 등등 한중일양 다 들어있도록 다양하게) 음식리스트로 해보기
