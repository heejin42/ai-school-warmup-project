## 9.10(목)
### 내용
- 사용자 기준으로 생각해보기  

최재현 : 사용자가 서비스를 이용할 때 입력 재료 데이터를 받아서 저장 후 그 데이터로 추천해주는 기능  

유세빈 : 사용자의 냉장고 재료들을 싹 받아서 추천. 추가로 유통기한 정보도 받아서 이것을 이용해서 추천. 디폴트 값으론 대략적 유통기한 잡아주고, 유통기한 지난 재료들이 있다면 경고 표시. 유통기한 가까운 재료를 이용해서 추천도 함   
=> 식재료 쇼핑을 할 경우 연동까지 시켜서 바로바로 재료들을 입력될 수 있게 한다면 Best!  

서예지 : 음성인식과 같은 방법으로 레시피를 요리 진행상황에 맞게 알려 준다. -> 이미 존재..ㅜ.ㅜ  

이희진 : 고객들의 정보(입맛에 따른 0 ~ 10까지)를 받아서 클러스터링 해보자. 음식이. 설문조사 : 본인 입맛과, 집에서 해먹어보고 싶은 요리, 자주 해먹는 요리. 비슷한 입맛의 다른 사용자들의 해먹은 요리들을 추천받을 수 있다.

    ex) 입맛 - 매운 맛의 음식을 좋아한다 (0 ~ 10), 단 맛의 음식을 좋아한다 (0~10), 짠 맛의 음식을 좋아한다 (0 ~ 10), 육류를 좋아한다 (0 ~ 10), 채소류를 좋아한다 (0 ~ 10), 해산물을 좋아한다 (0 ~ 10)   


- 기존 기술 보완  

서예지 : 리뷰 데이터를 자연어 분석하여 토픽을 추출해서 그것을 특성으로 삼아 라벨링 후, 라벨을 이용하여 각 요리의 내용을 토픽 분석
