```
# 감기
df = final_data[final_data['재료'].str.contains('우유|초콜렛|초콜릿|사과|라임')==False]
df = df[df['재료'].str.contains('콩나물|더덕|모과|무|배|도라지|칡|생강|유자|호박|대추|오렌지|포도|레몬|녹차|꿀|딸기|귤|피망|고추|브로콜리|굴')]
df = df[df['요리이름'].str.contains('차|스프|죽|미음')]
df_cold = df.sort_values(by=['비타민c', '단백질'], ascending=[False, False])

# 비만
df = final_data[final_data['태그'].str.contains('다이어트')]
df = df[df['재료'].str.contains('설탕|버터|마가린|쇠고기|소고기|돼지고기|스팸|흰밥|빵|떡')==False]
df = df[df['재료'].str.contains('다시마|미역|닭가슴살|아보카도|생강|대추|콩|자몽')]
df = df[df['요리이름'].str.contains('과자|쿠키|빵|떡')==False]
df_fatness = df.sort_values(by=['탄수화물','나트륨','당류'], ascending=[True, True, True])

# 빈혈
df = final_data[final_data['태그'].str.contains('빈혈')]
df = df[df['재료'].str.contains('밀가루|녹차|커피|초콜릿|초콜렛')==False]
df = df[df['재료'].str.contains('시금치|땅콩|소고기|쇠고기|노른자|미역|브로콜리|레드비트|굴|토마토|체리|연어')]
df_beenblood = df.sort_values(by=['비타민c', '단백질'], ascending=[False, False])

# 고혈압
df = final_data[final_data['재료'].str.contains('빵|밀가루|버터|마가린|치즈|햄|조개|마요네즈|새우|김치')==False]
df = df[df['재료'].str.contains('양파|마늘|시금치|바나나|계피|올리브오일|비트|귀리')]
df_toohigh = df.sort_values(by=['칼슘','비타민c','식이섬유'], ascending=[False, False, False])
# df_toohigh = df.sort_values(by=['나트륨','탄수화물','지질'], ascending=[True, True, True])

# 심부전
df = final_data[final_data['재료'].str.contains('고추|겨자|카레|우엉|파|부추|고구마|밀가루')==False]
df = df[df['재료'].str.contains('두부|바나나|콩|호두|요거트')]
df = df[df['요리이름'].str.contains('튀김|차|도넛')==False]
df_heartattack = final_data.sort_values(by=['콜레스테롤', '지질', '나트륨','단백질', '식이섬유'], ascending=[True, True, True, False, False])

# 고지혈증
df = final_data[final_data['재료'].str.contains('기름|마가린|빵|커피|베이컨|소시지|소세지|햄|치즈')==False]
df = df[df['재료'].str.contains('양파|마늘|파|토마토|호두|아몬드|카레')]
df = df[df['요리이름'].str.contains('라면|과자|아이스크림|빵|케이크')==False]
df_gojiblood = final_data.sort_values(by=['비타민c','식이섬유', '탄수화물','당류','콜레스테롤'], ascending=[False,False, True, True, True])

# 당뇨
df = final_data[final_data['재료'].str.contains('설탕|밀가루|조청|치즈|흰|버터')==False]
df = df[df['재료'].str.contains('현미|시금치|콩|두부|순두부|토마토|양파|대추|표고버섯|당근|녹두|고구마|옥수수|청국장|호두|둥글레|두릅|마늘|팥')]
df = df[df['요리이름'].str.contains('라면|과자|아이스크림|빵|케이크|젓갈|장아찌|주스|쥬스|튀김|조림')==False]
df_toosweet = df.sort_values(by=['식이섬유', '탄수화물','당류'], ascending=[False, True, True])

# 심근경색
df = final_data[final_data['재료'].str.contains('돼지갈비|소갈비|등심|안심|삼겹살|곱창|내장|햄|소시지|베이컨|버터|마요네즈|우유|생크림|치즈|밀가루|새우|장어|미꾸라지|오징어')==False]
df = df[df['재료'].str.contains('고등어|연어|참치|동태|갈치|두부|호두|아몬드|땅콩|콩')]
df = df[df['요리이름'].str.contains('튀김|도넛')==False]
df_heartproblem = df.sort_values(by=['식이섬유','나트륨', '지질','콜레스테롤'], ascending=[False, True, True, True])

# 뇌졸중
df = final_data[final_data['재료'].str.contains('오징어|새우|게|소시지|햄|베이컨|초콜릿|밀가루|치즈')==False]
df = df[df['재료'].str.contains('고등어|토마토|당근|사과|배|베리|아보카도|정어리|멜론|수박|녹차|귀리|생강|꽁치|부추|미역|다시마|연어|매생이')]
df_brainattack = df.sort_values(by=['식이섬유','비타민c','나트륨'], ascending=[False,False, True])

# 골다공증
df = final_data[final_data['재료'].str.contains('설탕|소금|커피')==False]
df = df[df['재료'].str.contains('우유|요거트|요구르트|고등어|연어|표고버섯|멸치|사과|귤|콩|콩나물|순두부|두부|아몬드|냉이|고사리|김|다시마|가오리|명태')]
df_boneattack = df.sort_values(by=['칼슘', '비타민c','나트륨'], ascending=[False, False, True])


def diseases(disease):

  if disease=="감기":
    return df_cold

  elif disease=="비만":
    return df_fatness

  elif disease=="빈혈":
    return df_beenblood

  elif disease=="고혈압":
    return df_toohigh

  elif disease=="심부전":
    return df_heartattack

  elif disease=="고지혈증":
    return df_gojiblood

  elif disease=="당뇨":
    return df_toosweet

  elif disease=="심근경색":
    return df_heartproblem

  elif disease=="뇌졸중":
    return df_brainattack

  elif disease=="골다공증":
    return df_boneattack

  else:
    return False

print('감기 | 비만 | 빈혈 | 고혈압 | 심부전 | 고지혈증 | 당뇨 | 심근경색 | 뇌졸중 | 골다공증')
print("=====================================================================================")
disease = input("질병을 입력하세요: ")
diseases(disease)
```
